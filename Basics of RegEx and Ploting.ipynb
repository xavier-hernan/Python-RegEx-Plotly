## Part 1 (regular expressions)

The goal is to extract dates of different formats from medical data.
We should correctly identify all of the different date variants encoded in this dataset and to properly standardize and sort the dates.

### Data loading

```python
with open("./medical_dataset.txt") as f:
    lines = f.readlines()

lines
```

###String vectorization

```python
import pandas as pd
pd.options.display.max_rows = None
pd.options.display.max_colwidth = None
df = pd.DataFrame(lines, columns=["text"])
df[:500]
```

### **DESIGNING ALL DATE FORMATS**

#### **FIRST FORMAT**
Here I take the date format with dd/mm/yyyy. Day can be 1 or 2 digits,  year can be 2 to 4 digits and month has to be writted in leters and have, at leats, the first three leters of the month. Besides, the separation between them can be space, / or -.
And, I take the date format with mm/yyyy. Month has to be writted in leters and have, at leats, the first three leters of the month and year has to be 4 digits. Besides, the separation between them can be space, '.', ',', '/', '-', two of them or nothing.

```python
df_aux = df["text"].str.extract(r'(?P<day>\d{0,2})[\s|/|-]?(?P<month>Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\w*[\s|/|-](?P<year>\d{4})').dropna(how = 'any')
df_aux
```

#### **SECOND FORMAT**
Here I take the date format with mm/dd/yyyy. Both month and day can be 1 or 2 digits, and year can be 2 to 4 digits. Besides, the separation between them can be space, / or -.
Also, I take the date format with mm/yyyy. Month can be 1 or 2 digits and year can be 2 to 4 digits. Besides, the separation between them can be space, / or -. 

```python
df_aux2 = df["text"].str.extract(r'(?P<month>\b\d{1,2})[/|-]{0,1}(?P<day>\b\d{0,2}\b)[\/|-]{0,1}(?P<year>\d{2}\b|\d{4})').dropna(how = 'any')
df_aux2
```

#### **THIRD FORMAT**
Here I take the date format with mm/dd/yyyy. Month has to be writted in leters and have, at leats, the first three leters of the month, day can be 1 or 2 digits and year has to be 4 digits. Besides, the separation between them can be space, '.', ',', '/', '-', two of them or nothing.

```python
df_aux3 = df["text"].str.extract(r'(?P<month>Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\w*[.|,|\s|/|-]{0,2}(?P<day>\d{0,2})[,|\s|/|-]{0,2}(?P<year>\d\d\d\d)').dropna(how = 'any')
df_aux3
```

#### **FOURTH FORMAT**
This one is not adding any value to our dataset. I have added it just in case other dataset include any format showing only the year, but it is not happening in our data.

```python
df_aux4 = df["text"].str.extract(r'(?P<year>\d{2,4})').dropna(how = 'any')
day = [1]* df_aux4.shape[0]
month = [1]* df_aux4.shape[0]
df_aux4['day'] = day
df_aux4['month'] = month
df_aux4
```

### **UNIFYING ALL DATE FORMATS IN ONE SINGLE DATAFRAME, WITHOUT DUPLICATES**

```python
import numpy as np
new_dataframe = pd.concat([df_aux, df_aux2, df_aux3, df_aux4])
new_dataframe = new_dataframe[~new_dataframe.index.duplicated(keep='first')].sort_index()
new_dataframe.replace('', np.nan, inplace=True)
new_dataframe['day'].fillna(1, inplace = True)
new_dataframe
```

Changing the strings by values in the month side. 

```python
months ={'Jan' : 1, 'Feb' : 2, 'Mar' : 3, 'Apr' : 4, 'May' : 5, 'Jun' : 6, 'Jul' : 7, 'Aug' : 8, 'Sep' : 9, 'Oct' : 10, 'Nov' : 11, 'Dec' : 12}
```

```python
df_final = new_dataframe.replace({'month': months})
df_final.loc[((df_final["year"].astype('int') <= 99) & (1 <= df_final["year"].astype('int'))), 
             'year'] = '19'+ df_final["year"].astype(str)
df_final
```

Finally, I add the text column to the dataframe.

```python
cols = ["year", "month", "day"]
df.loc[df_final.index, cols] = df_final[cols]

df
```

Here I am changing all the string formats for numbers.

```python
df.isna().sum()
```

Saving the dataframe to as excel file. (numbers have been set as int)

```python
df = df.astype({'day':'int', 'year':'int', 'month':'int'})
processed_dates = pd.ExcelWriter('processed_dates.xlsx')
df.to_excel(processed_dates)
processed_dates.save()
print('Dataframe has been saved to an excel file called "processed_dates.xlsx"')
```

##Part 2 (plotly)

The goal of this part is to explore using plotly / plotly express the following questions:

* Is there any relation between neighborhood and price?

* Is there any relation between neighborhood and year built?

* How overall quality, lot area, year built and price interact with each other?

* How quality, lot area, year built and price interact with each other and evolve in time?

```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
import plotly.express as px
py.init_notebook_mode(connected=True) #this allows to display plotly graphs in Jupyter
```

```python
df2 = pd.read_csv("./dataset_housing.csv")
df2
```

### 1. Is there any relation between neighborhood and price?

As we can see here, there is a clear relationship between neighborhood and price. Depending on the neighborhood, the houses there are cheper or expensive. For sure neighborhood determines part of the price.

For the neighborhoods that are at the left part of the graph the prices are lower, so we can guess they are bad neighborhoods.

For the neighborhoods that are at the right part of the graph, the price is higher, even the cheapest house at 'NridgHt' is more expensive than the best house at 'MeadowV'. So we can guess they are the best neighborhoods.

```python
fig = px.box(df2, x="Neighborhood", y="SalePrice", title='Relationship between Neighborhood and houses price')
fig.update_xaxes(categoryorder="median ascending")
py.iplot(fig)
```

### 2. Is there any relation between neighborhood and year built?

As we can see here, there is a clear relationship between neighborhood and year build. There are neighborhoods that only have new buildings, and others where any house is being constructed. 

From the graph we can guess that though time some neighborhoods have appeared, because some of them only have houses from year 2000. Also, there are other neighborhoods (like 'OldTown') where there are no houses constructed after 1980.

```python
fig2 = px.box(df2, x="Neighborhood", y="YearBuilt",title='Relationship between Neighborhood and the year its houses were build')
fig2.update_xaxes(categoryorder="median ascending")
py.iplot(fig2)
```

### 3.How overall quality, lot area, year built and price interact with each other?

At this graph we can see how overall quality, lot area, year built and price interact each other but only in pairs. What is the relationship between each one of them with all the others.

```python
fig3 = px.scatter_matrix(df2, dimensions=["OverallQual", "LotArea", "YearBuilt", "SalePrice"])
fig3.show()
```

Looking at this graph, we could say that the biggesr heouses are the more expensive ones. As we can appreciate the bubbles are getting bigger as they go higher in the graph.

Besides, the houses built in the recent years are the ones that has a higher Overall Quality. The colors for the houses at the right/top of the graph are more yellow colored than the ones that are at the left/down (purple ones). Also this applies for the sales price.

```python
fig4 = px.scatter(df2, x="YearBuilt", y="SalePrice", color="OverallQual",
                 size='LotArea', hover_data=['Id'])
fig4.show()
```

### 4. How quality, lot area, year built and price interact with each other and evolve in time?

Here we can see interactively how the previous interpretation evolve depending on the year hauses were sold.

```python
df2_order = df2.sort_values(by=["YrSold"], ascending=True)

fig5 = px.scatter(df2_order, x="YearBuilt", y="SalePrice", color="OverallQual", animation_frame="YrSold",
                 size='LotArea', hover_data=['Id'])
fig5.show()
```
